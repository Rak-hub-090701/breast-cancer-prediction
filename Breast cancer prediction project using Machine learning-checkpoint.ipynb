```python
#Main part of ML project spliting of dataset into independent(X) and dependent(Y) datasets
## .values extension in code gives output values in array
X = data.iloc[:,1:4].values
Y = data.iloc[:,5].values
```

```python
# X tells us features which detect whether patient has breast cancer or not
## Y tells patient has cancer or not
```

```python
#split the data set into 75% training and 25% testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25,random_state=0)
```

```python
#scale the data or called feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
  # create a function for the models
def models(X_train,Y_train):
    #using decision tree classifier
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy',random_state=0)
    tree.fit(X_train, Y_train)
    print('[0]Decision tree classifier accuracy is', tree.score(X_train,Y_train))
    
    #using logistic regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    print("[1]Logistic Regression accuracy is", log.score(X_train,Y_train))
     
    #using Random forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    forest.fit(X_train,Y_train)
    print("[2]Random forest classifier accuracy is", forest.score(X_train,Y_train))
    
    return tree, log, forest


  
    
```

```python
#Getting all of the models accuracy
model = models(X_train,Y_train)
```

```python
# test the models accuracy with confusion matrix
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print("Model", i)
    cm = confusion_matrix(Y_test, model[i].predict(X_test))
    
    TP = cm[0][0] #Truth positive
    TN = cm[1][1]
    FN = cm[1][0]
    FP = cm[0][1]
    print(cm)
    print("Testing accuracy = ", (TP + TN)/(TP + TN + FN + FP))
    
```

```python
#another way to get metrics of the models
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(model)):
    print("Model", i)
    print(classification_report(Y_test, model[i].predict(X_test)))
    print(accuracy_score(Y_test, model[i].predict(X_test)))
```
